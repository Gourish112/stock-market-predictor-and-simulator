```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
import yfinance as yf
import time
import os
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Dense, BatchNormalization, Conv1D, InputLayer, Bidirectional
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Step 1: Load Stock Market Data with error handling
try:
    start = dt.datetime(2010, 1, 1)
    end = dt.datetime.today()
    
    # Download data with retry mechanism
    max_attempts = 3
    for attempt in range(max_attempts):
        try:
            df = yf.download(tickers=['^GSPC'], start=start, end=end, auto_adjust=True)
            break
        except Exception as e:
            if attempt < max_attempts - 1:
                print(f"Attempt {attempt+1} failed. Retrying...")
                time.sleep(2)  # Wait before retry
            else:
                raise Exception(f"Failed to download data after {max_attempts} attempts: {e}")
    
    # Check if data is empty
    if df.empty:
        raise Exception("Downloaded dataframe is empty")
        
    # Drop columns if they exist
    df.drop(columns=['Adj Close'], inplace=True, errors='ignore')
    
    # Handle missing values appropriately
    print(f"Missing values before cleaning: {df.isna().sum().sum()}")
    df = df.dropna().reset_index()  # Remove missing values
    print(f"Missing values after cleaning: {df.isna().sum().sum()}")
    print(f"Total data points: {len(df)}")
    
except Exception as e:
    print(f"Error in data loading: {e}")
    exit(1)

# Step 2: Feature Engineering
try:
    # Calculate technical indicators
    # Moving averages
    df['MA5'] = df['Close'].rolling(window=5).mean()
    df['MA20'] = df['Close'].rolling(window=20).mean()
    df['MA50'] = df['Close'].rolling(window=50).mean()
    
    # MACD
    df['EMA12'] = df['Close'].ewm(span=12, adjust=False).mean()
    df['EMA26'] = df['Close'].ewm(span=26, adjust=False).mean()
    df['MACD'] = df['EMA12'] - df['EMA26']
    df['Signal'] = df['MACD'].ewm(span=9, adjust=False).mean()
    
    # RSI (14 period)
    delta = df['Close'].diff()
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)
    avg_gain = gain.rolling(window=14).mean()
    avg_loss = loss.rolling(window=14).mean()
    rs = avg_gain / avg_loss
    df['RSI'] = 100 - (100 / (1 + rs))
    
    # Bollinger Bands
    df['20d_std'] = df['Close'].rolling(window=20).std()
    df['upper_band'] = df['MA20'] + (df['20d_std'] * 2)
    df['lower_band'] = df['MA20'] - (df['20d_std'] * 2)
    
    # Price changes and volatility
    df['return_1d'] = df['Close'].pct_change()
    df['volatility_14d'] = df['return_1d'].rolling(window=14).std()
    
    # Day of week (market behavior differs by day)
    df['day_of_week'] = pd.to_datetime(df['Date']).dt.dayofweek
    
    # Month (seasonal effects)
    df['month'] = pd.to_datetime(df['Date']).dt.month
    
    # One-hot encode categorical features before dropping NAs
    for i in range(7):  # Days of week (0-6)
        df[f'day_{i}'] = (df['day_of_week'] == i).astype(int)
    
    for i in range(1, 13):  # Months (1-12)
        df[f'month_{i}'] = (df['month'] == i).astype(int)
    
    # Define feature columns
    feature_columns = ['Open', 'High', 'Low', 'Close', 'Volume', 
                      'MA5', 'MA20', 'MA50', 'MACD', 'Signal', 'RSI',
                      'upper_band', 'lower_band', 'return_1d', 'volatility_14d']
    
    # Add one-hot encoded columns to features
    for i in range(7):
        feature_columns.append(f'day_{i}')
    
    for i in range(1, 13):
        feature_columns.append(f'month_{i}')
    
    # Drop rows with NaN values (from the technical indicators calculations)
    df = df.dropna().reset_index(drop=True)
    print(f"Data shape after feature engineering: {df.shape}")
    
except Exception as e:
    print(f"Error in feature engineering: {e}")
    exit(1)

# Step 3: Train-Test-Validation Split (chronological)
try:
    # Proper chronological split for time series data
    train_size = int(len(df) * 0.7)
    val_size = int(len(df) * 0.15)
    
    data_training = df.iloc[:train_size]
    data_validation = df.iloc[train_size:train_size+val_size]
    data_testing = df.iloc[train_size+val_size:]
    
    print(f"Training set: {len(data_training)} rows")
    print(f"Validation set: {len(data_validation)} rows")
    print(f"Testing set: {len(data_testing)} rows")
    
except Exception as e:
    print(f"Error in data splitting: {e}")
    exit(1)

# Step 4: Normalization with separate scalers for features and target
try:    
    # Feature scaler
    feature_scaler = MinMaxScaler(feature_range=(0, 1))
    feature_scaler.fit(data_training[feature_columns])
    
    # Target scaler (specifically for Close price)
    target_scaler = MinMaxScaler(feature_range=(0, 1))
    target_scaler.fit(data_training[['Close']])
    
    # Transform datasets
    scaled_train_features = feature_scaler.transform(data_training[feature_columns])
    scaled_val_features = feature_scaler.transform(data_validation[feature_columns])
    scaled_test_features = feature_scaler.transform(data_testing[feature_columns])
    
    # Transform Close prices for targets
    scaled_train_close = target_scaler.transform(data_training[['Close']])
    scaled_val_close = target_scaler.transform(data_validation[['Close']])
    scaled_test_close = target_scaler.transform(data_testing[['Close']])
    
except Exception as e:
    print(f"Error in normalization: {e}")
    exit(1)

# Step 5: Create sequences with proper handling of prediction horizons
try:
    def create_sequences(features, targets, seq_length, horizons=[1, 5, 20, 60]):
        """
        Create sequences for multi-horizon forecasting with safety checks
        """
        X, y = [], []
        max_horizon = max(horizons)
        
        # Only create sequences where we have enough future data for all horizons
        for i in range(seq_length, len(features) - max_horizon):
            X.append(features[i-seq_length:i])
            y_values = [targets[i+h-1, 0] for h in horizons]  # -1 because horizons are 1-based
            y.append(y_values)
            
        return np.array(X), np.array(y)
    
    # Parameters
    sequence_length = 60  # 60 trading days (about 3 months)
    prediction_horizons = [1, 5, 20, 60]  # 1 day, 1 week, 1 month, 3 months
    horizon_names = ['1 day', '1 week', '1 month', '3 months']
    
    # Create sequences
    X_train, y_train = create_sequences(scaled_train_features, scaled_train_close, 
                                        sequence_length, prediction_horizons)
    
    X_val, y_val = create_sequences(scaled_val_features, scaled_val_close, 
                                   sequence_length, prediction_horizons)
    
    X_test, y_test = create_sequences(scaled_test_features, scaled_test_close, 
                                     sequence_length, prediction_horizons)
    
    print(f"Training sequences: {X_train.shape}")
    print(f"Validation sequences: {X_val.shape}")
    print(f"Testing sequences: {X_test.shape}")
    
except Exception as e:
    print(f"Error in sequence creation: {e}")
    exit(1)

# Step 6: Define Advanced Multi-Horizon LSTM Model
try:
    # Define model with bidirectional layers
    model = Sequential()
    
    # Input shape will be [sequence_length, num_features]
    model.add(InputLayer(input_shape=(X_train.shape[1], X_train.shape[2])))
    
    # CNN for feature extraction
    model.add(Conv1D(filters=64, kernel_size=3, padding='same', activation='relu'))
    
    # Bidirectional LSTM layers
    model.add(Bidirectional(
        LSTM(128, return_sequences=True, recurrent_dropout=0.1)
    ))
    model.add(BatchNormalization())
    model.add(Dropout(0.2))
    
    model.add(Bidirectional(
        LSTM(128, return_sequences=True, recurrent_dropout=0.1)
    ))
    model.add(BatchNormalization())
    model.add(Dropout(0.2))
    
    model.add(Bidirectional(
        LSTM(64, return_sequences=False)
    ))
    model.add(BatchNormalization())
    model.add(Dropout(0.2))
    
    # Output layer for multi-horizon forecasting
    model.add(Dense(len(prediction_horizons), activation='linear'))
    
    # Print model summary
    model.summary()
    
except Exception as e:
    print(f"Error in model definition: {e}")
    exit(1)

# Step 7: Compile & Train with advanced callbacks
try:
    # Create model directory if it doesn't exist
    model_dir = "stock_models"
    os.makedirs(model_dir, exist_ok=True)
    
    # Callbacks for better training
    callbacks = [
        # Stop training when validation loss doesn't improve
        EarlyStopping(
            monitor='val_loss',
            patience=20,
            restore_best_weights=True,
            verbose=1
        ),
        # Reduce learning rate when validation loss plateaus
        ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=5,
            min_lr=1e-6,
            verbose=1
        ),
        # Save best model
        ModelCheckpoint(
            filepath=os.path.join(model_dir, 'best_model.keras'),  # Fixed to .keras extension
            monitor='val_loss',
            save_best_only=True,
            verbose=1
        )
    ]
    
    # Compile model with Huber loss for robustness to outliers
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
        loss=tf.keras.losses.Huber(),
        metrics=['mae']
    )
    
    # Train model
    history = model.fit(
        X_train, y_train,
        epochs=100,
        batch_size=32,
        validation_data=(X_val, y_val),
        callbacks=callbacks,
        verbose=1
    )
    
    # Save final model
    model.save(os.path.join(model_dir, "final_stock_model.keras"))  # Fixed to .keras extension
    
except Exception as e:
    print(f"Error in model training: {e}")
    exit(1)

# Step 8: Visualize training history
try:
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(history.history['mae'], label='Training MAE')
    plt.plot(history.history['val_mae'], label='Validation MAE')
    plt.title('Model MAE')
    plt.xlabel('Epoch')
    plt.ylabel('MAE')
    plt.legend()
    
    plt.tight_layout()
    plt.savefig(os.path.join(model_dir, 'training_history.png'))
    plt.show()
    
except Exception as e:
    print(f"Error in history visualization: {e}")

# Step 9: Model Performance Evaluation on Test Data
try:
    # Predict on test data
    y_pred = model.predict(X_test)
    
    # Initialize arrays for storing rescaled values
    y_test_rescaled = np.zeros_like(y_test)
    y_pred_rescaled = np.zeros_like(y_pred)
    
    # Inverse transform predictions & actual values
    for i in range(len(prediction_horizons)):
        y_test_rescaled[:, i] = target_scaler.inverse_transform(y_test[:, i].reshape(-1, 1)).flatten()
        y_pred_rescaled[:, i] = target_scaler.inverse_transform(y_pred[:, i].reshape(-1, 1)).flatten()
    
    # Compute Performance Metrics for each horizon
    print("\n===== MODEL PERFORMANCE =====")
    for i, horizon in enumerate(horizon_names):
        mse = mean_squared_error(y_test_rescaled[:, i], y_pred_rescaled[:, i])
        rmse = np.sqrt(mse)
        mae = mean_absolute_error(y_test_rescaled[:, i], y_pred_rescaled[:, i])
        mape = np.mean(np.abs((y_test_rescaled[:, i] - y_pred_rescaled[:, i]) / y_test_rescaled[:, i])) * 100
        r2 = r2_score(y_test_rescaled[:, i], y_pred_rescaled[:, i])
        
        print(f"\n🔹 Forecast Horizon: {horizon}")
        print(f"   • Mean Squared Error (MSE): {mse:.4f}")
        print(f"   • Root Mean Squared Error (RMSE): {rmse:.4f}")
        print(f"   • Mean Absolute Error (MAE): {mae:.4f}")
        print(f"   • Mean Absolute Percentage Error (MAPE): {mape:.2f}")
        print(f"   • R² Score: {r2:.4f}")
except Exception as e:
    print(f"Error in model evaluation: {e}")

# Step 10: Plot Actual vs Predicted Prices for each horizon
try:
    plt.figure(figsize=(20, 15))
    
    for i, horizon in enumerate(horizon_names):
        plt.subplot(2, 2, i+1)
        
        # Plot actual values
        plt.plot(y_test_rescaled[:, i], label="Actual Price", color='blue', alpha=0.7)
        
        # Plot predicted values
        plt.plot(y_pred_rescaled[:, i], label="Predicted Price", color='red', linestyle="dashed")
        
        # Calculate error bounds (prediction intervals)
        mae_value = mean_absolute_error(y_test_rescaled[:, i], y_pred_rescaled[:, i])
        plt.fill_between(
            range(len(y_pred_rescaled[:, i])),
            y_pred_rescaled[:, i] - mae_value*1.96,
            y_pred_rescaled[:, i] + mae_value*1.96,
            color='red', alpha=0.1, label='95% Confidence'
        )
        
        plt.title(f"Stock Price Prediction ({horizon} Ahead)")
        plt.xlabel("Trading Days")
        plt.ylabel("Price ($)")
        plt.legend()
        plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig(os.path.join(model_dir, 'prediction_results.png'))
    plt.show()
    
except Exception as e:
    print(f"Error in plotting results: {e}")

# Step 11: Feature Importance Analysis
try:
    # Permutation Feature Importance
    from sklearn.inspection import permutation_importance
    
    # Flatten the 3D input for permutation importance
    # This is a simplification for feature importance analysis
    X_test_flat = X_test.reshape(X_test.shape[0], -1)
    
    # Train a simple model on flattened data for feature importance
    simple_model = tf.keras.Sequential([
        tf.keras.layers.Dense(64, activation='relu', input_shape=(X_test_flat.shape[1],)),
        tf.keras.layers.Dense(32, activation='relu'),
        tf.keras.layers.Dense(len(prediction_horizons))
    ])
    
    simple_model.compile(optimizer='adam', loss='mse')
    simple_model.fit(X_test_flat, y_test, epochs=5, verbose=0)
    
    # Perform permutation importance
    r = permutation_importance(simple_model, X_test_flat, y_test, 
                              n_repeats=10, random_state=42, n_jobs=-1)
    
    # Get feature importance scores
    importances = r.importances_mean
    
    # Reshape importances to match original feature structure
    feature_importances = {}
    for feature_idx, feature_name in enumerate(feature_columns):
        # For each feature, sum importance across time steps
        importance_sum = 0
        for i in range(sequence_length):
            flat_idx = i * len(feature_columns) + feature_idx
            if flat_idx < len(importances):
                importance_sum += importances[flat_idx]
        feature_importances[feature_name] = importance_sum
    
    # Sort features by importance
    sorted_importances = sorted(feature_importances.items(), key=lambda x: x[1], reverse=True)
    
    # Plot top 15 feature importances
    plt.figure(figsize=(12, 8))
    top_features = [item[0] for item in sorted_importances[:15]]
    top_scores = [item[1] for item in sorted_importances[:15]]
    
    plt.barh(top_features, top_scores)
    plt.xlabel('Importance Score')
    plt.ylabel('Feature')
    plt.title('Top 15 Feature Importance')
    plt.tight_layout()
    plt.savefig(os.path.join(model_dir, 'feature_importance.png'))
    plt.show()
    
except Exception as e:
    print(f"Error in feature importance analysis: {e}")
    print(f"Detailed error: {str(e)}")

# Step 12: Make future predictions
try:
    # Get the most recent data for prediction
    recent_data = df.iloc[-sequence_length:][feature_columns].values
    recent_scaled = feature_scaler.transform(recent_data)
    
    # Reshape for model input (adding batch dimension)
    model_input = recent_scaled.reshape(1, sequence_length, len(feature_columns))
    
    # Make predictions
    future_predictions = model.predict(model_input)
    
    # Convert predictions back to original scale
    future_predictions_rescaled = np.zeros((1, len(prediction_horizons)))
    for i in range(len(prediction_horizons)):
        future_predictions_rescaled[0, i] = target_scaler.inverse_transform(
            future_predictions[0, i].reshape(-1, 1)
        )[0, 0]
    
    # Get the last known closing price
    last_price = df['Close'].iloc[-1]
    last_date = df['Date'].iloc[-1]
    
    # Print predictions
    print("\n===== FUTURE PREDICTIONS =====")
    print(f"Last known price (as of {last_date}): ${last_price:.2f}")
    
    for i, horizon in enumerate(horizon_names):
        print(f"Predicted price {horizon} ahead: ${future_predictions_rescaled[0, i]:.2f}")
    
    # Plot predictions
    plt.figure(figsize=(12, 6))
    
    # Historical data (last 60 days)
    historical_dates = df['Date'].iloc[-60:].values
    historical_prices = df['Close'].iloc[-60:].values
    
    # Get dates for future predictions
    from pandas.tseries.offsets import BDay  # Business day
    
    future_dates = [
        pd.to_datetime(last_date) + BDay(horizon)
        for horizon in prediction_horizons
    ]
    
    # Plot historical and future predictions
    plt.plot(historical_dates, historical_prices, label='Historical Prices', color='blue')
    
    # Highlight the last known price
    plt.scatter([historical_dates[-1]], [historical_prices[-1]], color='blue', s=100, zorder=5)
    
    # Plot predictions
    plt.scatter(
        future_dates, 
        future_predictions_rescaled[0], 
        label='Predictions', 
        color='red', 
        marker='*', 
        s=200,
        zorder=5
    )
    
    # Connect with a dashed line
    plt.plot(future_dates, future_predictions_rescaled[0], 'r--', alpha=0.7)
    
    plt.title('Stock Price Forecast')
    plt.xlabel('Date')
    plt.ylabel('Price ($)')
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig(os.path.join(model_dir, 'future_predictions.png'))
    plt.show()
    
except Exception as e:
    print(f"Error in future predictions: {e}")
    print(f"Detailed error: {str(e)}")

# Step 13: Add a prediction confidence metric
try:
    # Calculate prediction confidence based on historical error
    print("\n===== PREDICTION CONFIDENCE =====")
    
    for i, horizon in enumerate(horizon_names):
        # Calculate mean absolute percentage error for this horizon
        mape = np.mean(np.abs((y_test_rescaled[:, i] - y_pred_rescaled[:, i]) / y_test_rescaled[:, i])) * 100
        
        # Convert MAPE to a confidence score (100% - MAPE, with a minimum of 0)
        confidence = max(0, 100 - mape)
        
        # Define confidence level
        if confidence >= 90:
            confidence_level = "Very High"
        elif confidence >= 80:
            confidence_level = "High"
        elif confidence >= 70:
            confidence_level = "Moderate"
        elif confidence >= 60:
            confidence_level = "Low"
        else:
            confidence_level = "Very Low"
            
        print(f"{horizon} prediction confidence: {confidence:.1f}% ({confidence_level})")
        
    print("\nNote: Confidence is based on historical prediction accuracy")
    
except Exception as e:
    print(f"Error in confidence calculation: {e}")

print("\nModel training and evaluation complete. Check the 'stock_models' directory for saved models and visualizations.")
```
